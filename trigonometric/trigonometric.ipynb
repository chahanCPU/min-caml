```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def cos(x):
    pi = 3.141592654
    if x >= 0.0:
        if x <= pi:
            x2 = x * x
            return 1.0 - x2 * (0.5 - x2 * (0.04166666667 - x2 * (0.001388888889 - x2 * 0.00002480158730)))
        else:
            if x <= 2.0 * pi:
                return -cos(x - pi)
            else:
                return cos(x - 2.0 * pi)
    else:
        return cos(-x)
```

```python
plt.figure(figsize=(8, 6), dpi=80)
plt.subplot(1, 1, 1)

x = np.linspace(-4.0 * np.pi, 4.0 * np.pi, 100000, endpoint=True)
y, z = ((np.vectorize)(cos))(x), np.cos(x)

plt.plot(x, y, color="blue", linewidth=1.0, linestyle="-")
plt.plot(x, z, color="green", linewidth=1.0, linestyle="-")

plt.xlim(-4.0 * np.pi, 4.0 * np.pi)
plt.xticks(np.linspace(-4.0 * np.pi, 4.0 * np.pi, 9, endpoint=True))
plt.ylim(-1.2, 1.2)
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))

plt.savefig("cos.png", dpi=72)
plt.show()
```

```python
def sin(x):
    pi = 3.141592654
    if x >= 0.0:
        if x <= pi:
            x2 = x * x
            return x * (1.0 - x2 * (0.1666666667 - x2 * (0.008333333333 - x2 * 0.0001984126984)))
        else:
            if x <= 2.0 * pi:
                return -sin(x - pi)
            else:
                return sin(x - 2.0 * pi)
    else:
        return -sin(-x)
```

```python
plt.figure(figsize=(8, 6), dpi=80)
plt.subplot(1, 1, 1)

x = np.linspace(-4.0 * np.pi, 4.0 * np.pi, 100000, endpoint=True)
y, z = ((np.vectorize)(sin))(x), np.sin(x)

plt.plot(x, y, color="blue", linewidth=1.0, linestyle="-")
plt.plot(x, z, color="green", linewidth=1.0, linestyle="-")

plt.xlim(-4.0 * np.pi, 4.0 * np.pi)
plt.xticks(np.linspace(-4.0 * np.pi, 4.0 * np.pi, 9, endpoint=True))
plt.ylim(-1.2, 1.2)
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))

plt.savefig("sin.png", dpi=72)
plt.show()
```

```python
def tan(x):
    pi = 3.141592654
    if x >= 0.0:
        if x <= pi / 2.0:
            x2 = x * x
            return x * (1.0 + x2 * (0.333333333 + x2 * (0.133333333 + x2 * 0.053968254)))
        else:
            return tan(x - pi)
    else:
        return -tan(-x)
```

```python
plt.figure(figsize=(8, 6), dpi=80)
plt.subplot(1, 1, 1)

x = np.linspace(-2.5 * np.pi, 2.5 * np.pi, 100000, endpoint=True)
y, z = ((np.vectorize)(tan))(x), np.tan(x)

plt.plot(x, y, color="blue", linewidth=1.0, linestyle="-")
plt.plot(x, z, color="green", linewidth=1.0, linestyle="-")

plt.xlim(-2.5 * np.pi, 2.5 * np.pi)
plt.xticks(np.linspace(-2.5 * np.pi, 2.5 * np.pi, 11, endpoint=True))
plt.ylim(-5.0, 5.0)
plt.yticks(np.linspace(-5, 5, 5, endpoint=True))

plt.savefig("tan.png", dpi=72)
plt.show()
```

```python
def atan(x):
    x2 = x * x
    return x * (1.0 - x2 * (0.3333333333 - x2 * (0.2 - x2 * 0.1428571429)))
```

```python
plt.figure(figsize=(8, 6), dpi=80)
plt.subplot(1, 1, 1)

x = np.linspace(-0.5 * np.pi, 0.5 * np.pi, 10000, endpoint=False)
y, z = ((np.vectorize)(atan))(x), ((np.vectorize)(math.atan))(x)

plt.plot(x, y, color="blue", linewidth=1.0, linestyle="-")
plt.plot(x, z, color="green", linewidth=1.0, linestyle="-")

plt.xlim(-0.5 * np.pi, 0.5 * np.pi)
plt.xticks(np.linspace(-0.5 * np.pi, 0.5 * np.pi, 9, endpoint=True))
plt.ylim(-1.2, 1.2)
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))

plt.savefig("atan.png", dpi=72)
plt.show()
```
